# Visualize samples generated as .npz

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
# load npz
file_name = '../assets/samples/imgnet_samples_500x64x64x3.npz'
npz_file = np.load(file_name)
samples = npz_file['arr_0']
```

```python
def viz_samples(samples):
    num_samples = samples.shape[0]
    img_shape = samples.shape[1:]
    print(num_samples, "number of samples of shape:", img_shape)

    num_cols = 5
    num_rows = (num_samples + num_cols - 1) // num_cols
    fig = plt.figure(figsize=(20, 4 * num_rows))
    gs = GridSpec(num_rows, num_cols, figure=fig)
    
    max_samples = 20
    samples = samples[:max_samples]

    for i, sample in enumerate(samples):
        row = i // num_cols
        col = i % num_cols
        ax = fig.add_subplot(gs[row, col])
        ax.imshow(sample, cmap='gray')
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
viz_samples(samples)
```
